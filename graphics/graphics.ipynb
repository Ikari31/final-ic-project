```python
import pandas as pd

rf_centralized = pd.read_csv('./graphics_csv/cpu_usage_and_energy_random_forest.csv')
rf_federated = pd.read_csv('./graphics_csv/federated_client_cpu_energy_log.csv')

display(rf_centralized)
display(rf_federated)
```

```python
import matplotlib.pyplot as plt

# Calcular a média do uso de CPU para cada dataset
cpu_usage_mean_centralized = rf_centralized['CPU_Usage(%)'].mean()
cpu_usage_mean_federated = rf_federated['CPU Usage (%)'].mean()

labels = ['Server', 'Client 1', 'Client 2', 'Centralized']
cpu_usage_means = [20, cpu_usage_mean_federated, cpu_usage_mean_federated, cpu_usage_mean_centralized]

# Criar gráfico de barras com as modificações solicitadas
plt.figure(figsize=(8, 6))
plt.bar(labels, cpu_usage_means, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('Devices', fontsize=16)
plt.ylabel('CPU Usage (%)',fontsize=16)

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calcular as médias de CPU_Time(s)
cpu_time_mean_centralized = rf_centralized['CPU_Time(s)'].mean()
cpu_time_mean_federated = rf_federated['Training Time (s)'].mean()

# Valores de CPU Time para os dispositivos, incluindo o 'Server' e duplicação do 'Federated Client'
labels = ['Server', 'Client 1', 'Client 2', 'Centralized']
cpu_time_means = [0.12, cpu_time_mean_federated, cpu_time_mean_federated, cpu_time_mean_centralized]  # Exemplo fictício para Server

# Criar gráfico de barras para CPU Time
plt.figure(figsize=(8, 6))
plt.bar(labels, cpu_time_means, color=['blue', 'green', 'orange', 'red'])

plt.xlabel('Devices', fontsize=16)
plt.ylabel('CPU Time (s)',fontsize=16)

# Mostrar gráfico de CPU Time
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calcular as médias de CPU_Energy(W)
cpu_energy_mean_centralized = rf_centralized['CPU_Energy(W)'].mean()
cpu_energy_mean_federated = rf_federated['Energy Consumed (W)'].mean()

# Valores de CPU Energy para os dispositivos, incluindo o 'Server' e duplicação do 'Federated Client'
labels = ['Server', 'Client 1', 'Client 2', 'Centralized']
cpu_energy_means = [1.5, cpu_energy_mean_federated, cpu_energy_mean_federated, cpu_energy_mean_centralized]  # Exemplo fictício para Server

# Criar gráfico de barras para CPU Energy
plt.figure(figsize=(8, 6))
plt.bar(labels, cpu_energy_means, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('Devices', fontsize=16)
plt.ylabel('CPU Energy (W)', fontsize=16)


# Mostrar gráfico de CPU Energy
plt.show()
```
